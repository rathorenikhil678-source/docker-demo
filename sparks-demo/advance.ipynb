```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StringType, IntegerType
from pyspark.sql.functions import udf

spark = SparkSession.builder \
    .appName("PartitionDemo") \
    .getOrCreate()

# Create sample data
data = [(i, i) for i in range(20)]


df = spark.createDataFrame(data, ["id", "value",])


print("Before Partitions:", df.rdd.getNumPartitions())


df_repartitioned = df.repartition(25)
print("After Partitions:", df_repartitioned.rdd.getNumPartitions())

# Coalesce to 2 partitions
df_coalesced = df_repartitioned.coalesce(10)
print("After Coalesced:", df_coalesced.rdd.getNumPartitions())
```
